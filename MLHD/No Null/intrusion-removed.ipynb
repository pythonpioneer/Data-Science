```python
import pandas as pd
import numpy as np
import os
global _path
_path = os.path.dirname(os.getcwd()) + '/horse data/horse-details.csv'
_path 
```

```python
# now read our data using pandas
hd = pd.read_csv(_path)
hd
```

```python
# now we want to take mean of horse-height and horse weight
height_mean = hd.dropna()['Horse-Height'].mean()
weight_mean = hd.dropna()['Horse-Weight'].mean()
print('height mean: ', height_mean)
'weight mean: ' + str(weight_mean)
```

```python
hd.head()
```

```python
no_null_hd = hd.dropna()
no_null_hd
```

```python
# now create a csv file with removing all null values
no_null_hd.to_csv('no-null-horse-details.csv')
```

```python
# now we will fill all the null values with average value
hd['Horse-Weight'].fillna(value=weight_mean, inplace=True)
hd['Horse-Height'].fillna(value=height_mean, inplace=True)
```
