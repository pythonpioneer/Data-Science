```python
# importing all requirements
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

algo = ['linear', 'ridge', 'lasso', 'poly', 'svm', 'dec-tre', 'ran-for']
```

```python
# cretaing data for residuals
data = {
    'MAE': [60.314924, 60.709698, 60.719184, 25.865330, 27.302353, 19.805842, 19.805842],
    'MSE': [6030.563693, 6094.940531, 6095.045622, 1096.266793, 1219.924728, 578.532270, 578.532433],
    'RMSE': [77.656704, 78.070100, 78.070773, 33.109919, 34.927420, 24.052697, 24.052701]
}
df = pd.DataFrame(data, index=algo)
df
```

```python
# now try to plot the graph for MAE
plt.subplots(figsize=(18, 8))
sns.barplot(x=algo, y='MAE', data=df)
```

```python
# now try to plot the graph for MSE
plt.subplots(figsize=(18, 8))
sns.barplot(x=algo, y='MSE', data=df)
```

```python
# now try to plot the graph for RMSE
plt.subplots(figsize=(18, 8))
sns.barplot(x=algo, y='RMSE', data=df)
```
